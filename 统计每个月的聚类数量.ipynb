```python
import pandas as pd
#df_cluster = pd.read_csv('som_good_cluster8result.csv',index_col = 0)
df_cluster = pd.read_csv('som_good_cluster8_0913result.csv',index_col = 0)
df_cluster['datetime'] = pd.to_datetime(df_cluster.index)
df_cluster.set_index("datetime", inplace=True)
df_cluster.info()
```

```python
df_cluster['day'] = pd.to_datetime(df_cluster.index).date
```

```python
df_cluster.head(3)
```

```python
df_cluster['YearMonth'] = df_cluster['day'].map(lambda x: 100*x.year + x.month)
```

```python
df_cluster.head(4)
```

```python
df_cluster = df_cluster['2018-01-01 00:00:00':]
```

```python
df_monthyear = df_cluster.groupby('YearMonth')['cluster'].value_counts().to_frame('每个月各个聚类计数').unstack()#reset_index()
```

```python
df_my = df_monthyear.fillna(0)
```

```python
df_my.to_csv('月统计.csv')
```

```python
df_my.head(3)
```

```python
frame=pd.DataFrame(df_my.values,index=df_my.index,
               columns=['cluster_0','cluster_1','cluster_2','cluster_3','cluster_4','cluster_5','cluster_6','cluster_7'])
```

```python
frame =frame.astype(int)
frame.head(3)
```

```python
import pyecharts
from pyecharts import Overlap, Bar, Line, Grid, EffectScatter
grid = Grid()
v1 = list( frame['cluster_0'].values)       
v2 = list(frame['cluster_1'].values)          
v3 = list(frame['cluster_2'].values) 
v4 = list(frame['cluster_3'].values)    
v5 = list(frame['cluster_4'].values)  
v6 = list(frame['cluster_5'].values)  
v7 = list(frame['cluster_6'].values) 
v8 = list(frame['cluster_7'].values) 
my_attr = ["第{}月".format(i) for i in range(1, len(v1)+1)] 
#[2.0, 2.2, 3.3, 4.5, 6.3, 10.2, 20.3, 23.4, 23.0, 16.5, 12.0, 6.2]
bar = Bar(title="", title_pos="10%")
bar.add("聚类0", my_attr, v1,legend_pos="55%",is_more_utils = True,)
bar.add("聚类1", my_attr,v2,yaxis_formatter=" 个/小时",legend_pos ='25%',
        legend_orient="vertical",legend_top="15%")
bar.add("聚类2", my_attr, v3)
bar.add("聚类3", my_attr, v4)
bar.add("聚类4", my_attr, v5)
bar.add("聚类5", my_attr, v6)
bar.add("聚类6", my_attr, v7)
bar.add("聚类7", my_attr, v8,yaxis_formatter=" 个/小时",legend_pos ='25%',
        legend_orient="vertical",legend_top="15%")
overlap = Overlap(width=1200, height=600)
overlap.add(bar)
bar
```
